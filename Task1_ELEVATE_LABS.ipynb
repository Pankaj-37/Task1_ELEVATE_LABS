```python
#checking for null values 

df.isnull().any()
```

```python
#dropping nam/missing values 
df.dropna(inplace=True)
```

```python
df.isnull().any().sum()
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
#Standardize text values 
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].str.strip().str.lower()

df
```

```python
#converting object/categorical columns which are in wrong format
df["date_added"]=pd.to_datetime(df['date_added'], format='%d-%m-%Y', errors='coerce')
```

```python
df['release_year']=pd.to_datetime(df['release_year'])
```

```python
df.info()
```

```python
# Clean and rename column headers
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df.head()
```

```python
# converting  duration column from strings like '125 min' to integers (e.g., 125)
df['duration_clean'] = df['duration'].str.extract('(\d+)')  # extracts the number part only
df['duration_clean'] = pd.to_numeric(df['duration_clean'], errors='coerce')  # convert to float (or int) and handle errors
df = df.drop("duration", axis=1)
```
